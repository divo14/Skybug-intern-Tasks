```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv('titanic.csv')
df
```

```python
df = df.drop(['boat','body','home.dest', 'sibsp','parch','ticket', 'cabin','embarked'], axis=1)
df
```

```python
target = df['survived']
target = target.fillna(0)
target
```

```python
inputs = df.drop(['name', 'survived'], axis=1)
inputs
```

```python
inputs.sex = inputs.sex.map({'male': 1, 'female': 2})
inputs.age = inputs.age.fillna(inputs.age.mean())
inputs.fare = inputs.fare.fillna(inputs.fare.mean())
inputs.sex = inputs.sex.fillna(1)
inputs.pclass = inputs.pclass.fillna(2)
inputs
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test, y_test)
```
